```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
plt.style.use('seaborn-whitegrid')
```

```python
clrs = sns.color_palette('husl', n_colors=3)  # a list of RGB tuples
ticks = [0.02, 0.04, 0.08]
labels = ['n=2', 'n=4', 'n=8']
n2 = [0.1013, 0.1142, 0.1875]
n4 = [0.0854, 0.0417, 0.0042]
n8 = [0.399, 0.1094, 0.0586]

cols = [n2, n4, n8]

plt.figure(figsize=(8,6))
ind = np.arange(3) 
width = 0.25
for i, col in enumerate(cols):
    plt.bar(ind + width * i, col, width, label=labels[i], color=clrs[i])

plt.title("Error Rate vs Seconds per Tone (distance=1 ft)")
plt.ylabel('Error Rate')

plt.xlabel('Length of Tone (s)')
plt.xticks(ind + width, ticks)
plt.ylim(0,1)

plt.legend(loc='best')

plt.show()
```
